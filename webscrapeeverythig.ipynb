Webscrapping NYTimes

```python
def getReq(site):
    climreq = requests.get(site)
    climhtml = climreq.text
    clim = BeautifulSoup(climhtml, "lxml")
    return clim
```

```python
climreq = requests.get("https://www.nytimes.com/pages/politics/index.html")
climhtml = climreq.text
clim = BeautifulSoup(climhtml, "lxml")
links = clim.find_all('div', {'class': ['storyHeader']})
h3 = clim.find_all('h3')
links.append(h3)
links = [val for sublist in links for val in sublist]
links = links[1:len(links)]
l = []
for i in range(0,len(links)):
    try:
        l.append((links[i].getText()).encode('ascii','ignore'))
        #l = l.translate({ 0x2018:0x27, 0x2019:0x27, 0x201C:0x22, 0x201D:0x22, 0x2026:0x20 })
    except:
        continue
links = map(lambda s: s.strip(), l)
links = map(lambda s: s.encode("utf-8"), links)
links = [x.rstrip('\n') for x in links]
```

```python
nytimes = links
```

Webscraping Fox News

```python
def fox_politics_titles(url):
    foxreq = requests.get(url)
    fox_bs = BeautifulSoup(foxreq.text,'lxml')
    h3_content = fox_bs.select("div h3")
    title_names = [x.text.strip().translate({ 0x2018:0x27, 0x2019:0x27, 0x201C:0x22, 0x201D:0x22, 0x2026:0x20 }).encode('ascii','ignore') for x in h3_content]
    title_names = set(title_names)
    return title_names
```

```python
url = "http://www.foxnews.com/politics.html"
title_names = fox_politics_titles(url)
```

```python
def extract_pol_titles(url):
    washreq = requests.get(url)
    wash_bs = BeautifulSoup(washreq.text,'lxml')
    a_content = wash_bs.select("div > h3 > a")
    title_names = [x.text.strip().translate({ 0x2018:0x27, 0x2019:0x27, 0x201C:0x22, 0x201D:0x22, 0x2026:0x20 }).encode('ascii','ignore') for x in a_content]
    title_names = set(title_names)
    return title_names
```

```python
trumps_admin = ['whitehouse','trump','conway','sessions','pence','president','tillerson','devos','flynn','kushner','carson','department','preibus','bannon','spicer','miller']
```

```python
frontpage = [x.lower() for x in frontpage]
abc = [x.lower() for x in abc]
wash = [x.lower() for x in wash]
fox= [x.lower() for x in fox]
c = [x.lower() for x in c]
wsj = [x.lower() for x in wsj]
npr = [x.lower() for x in npr]
economist = [x.lower() for x in economist]
nytimes = [x.lower() for x in nytimes]
```

```python
def trump_titles(title_names):
    """
    This function filters the article list with trump admin words
    Input: Unfiltered list of titles
    Output: Filtered list of titles
    """
    
    trumps_admin = ['whitehouse','Trump','conway','sessions','pence','president','tillerson','devos','flynn','kushner','carson','department','preibus','bannon','spicer','miller']

    total_titles = []

    for admin in trumps_admin:
        titles1 = [x for x in title_names if admin in x]
        total_titles.extend(titles1)

    return total_titles
```

```python
huff = trump_titles(frontpage)
abc = trump_titles(abc)
wash = trump_titles(wash)
fox = trump_titles(fox)
breitbart = trump_titles(c)
wsj = trump_titles(wsj)
npr = trump_titles(npr)
economist = trump_titles(economist)
nytimes = trump_titles(nytimes)
```
